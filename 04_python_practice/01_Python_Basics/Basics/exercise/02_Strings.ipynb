````python
# Middle char(s)

def middle(s):
    n = len(s)
    mid = n // 2
    if n % 2 == 1:
        return s[mid]
    else:
        return s[mid-1:mid+1]

assert middle("abc") == "b"
assert middle("abba") == "bb"
````

````python
# Title-case safely
def safe_title(s):
    words = s.split()
    result = []
    
    for word in words:
        if word.isupper():
            result.append(word)
        else:
            result.append(word.capitalize())
            
    return " ".join(result)

assert safe_title("learn SQL now")=="Learn SQL Now"
````

````python
# Reverse words
import re

def reverse_words(s):
    parts = re.split(r'(\s+)', s)
    parts[::2] = reversed(parts[::2])
    return "".join(parts)

assert reverse_words("one two  three") == "three two  one"
````

````python
# Find all indexes
def find_all(s, sub):
    result = []
    
    for i in range(len(s)):
        if s[i:i+len(sub)] == sub:
            result.append(i)
            
    return result

assert find_all("aaaa", "aa") == [0,1,2]
````

````python
# Anagram check
from collections import Counter

def is_anagram(a, b):
    return Counter(a.lower()) == Counter(b.lower())

assert is_anagram("Listen", "Silent")
````

````python
# Format table row
def fmt_row(values, widths):
    cols=[]
    
    for v, w in zip(values, widths):
        col = w - len(v)
        cols.append(v + " " * col)
        
    return " | ".join(cols)

assert fmt_row(["a","bb"], [3,4]) == "a   | bb  "
````

````python
# Slugify
def slugify(title):
    s = title.lower().strip()
    s = re.sub(r'[^a-z0-9]+', '-', s)
    s = s.strip('-')
    return s
    
assert slugify("Hello,  World!!") == "hello-world"
````
