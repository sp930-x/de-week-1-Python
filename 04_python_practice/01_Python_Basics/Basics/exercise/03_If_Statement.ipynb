````python
# FizzBuzz (classic)

def fizzbuzz(n):
    if n % 15 == 0:
        return "FizzBuzz"
    elif n % 3 == 0:
        return "Fizz"
    elif n % 5 == 0:
        return "Buzz"
    else:
        return str(n)

assert fizzbuzz(15) == "FizzBuzz"
````

````python
# Categorize temperature
def temp_label(c):
    if c < 0:
        return "freezing"
    elif c < 10:
        return "cold"
    elif c < 20:
        return "mild"
    elif c < 30:
        return "warm"
    else:
        return "hot"
        
assert temp_label(0) == "cold"
assert temp_label(10) == "mild"
assert temp_label(20) == "warm"
assert temp_label(30) == "hot"
````

````python
# Truthy length
def choose_label(s):
    s = s.strip()
    
    if len(s) == 0:
        return "empty"
    elif len(s) <= 3:
        return "short"
    else:
        return "long"

assert choose_label("") == "empty"
assert choose_label("   ") == "empty"
assert choose_label("hi") == "short"
````

````python
# Grade boundaries
def grade(score):
    if score < 0 or score > 100:
        raise ValueError("score must be between 0 and 100")

    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"
        
assert grade(95) == "A"
assert grade(85) == "B"
assert grade(75) == "C"
assert grade(65) == "D"
assert grade(50) == "F"
````

````python
# Password strength (simple)
def password_strength(s):
    has_letter = any(c.isalpha() for c in s)
    has_digit = any(c.isdigit() for c in s)
    has_special = any(not c.isalnum() for c in s)

    if len(s) < 6 or (s.isalpha() or s.isdigit()):
        return "weak"
    elif len(s) >= 8 and has_letter and has_digit and has_special:
        return "strong"
    elif len(s) >= 6 and has_letter and has_digit:
        return "medium"
    else:
        return "weak"

assert password_strength("abcdef") == "weak"
assert password_strength("abc123") == "medium"
assert password_strength("abc123!") == "medium"
assert password_strength("abc123!@") == "strong"
````

````python
def shipping_cost(weight_kg, country, premium=False, local_country="IN"):
    
   # * `shipping_cost(weight_kg, country, premium=False)`:

   #   * base by weight: ≤1kg: 5; ≤5kg: 10; else: 20
   #   * non-local country adds 15
   #   * premium adds +25 at the end
    if weight_kg <= 1:
        cost = 5
    elif weight_kg <= 5:
        cost = 10
    else:
        cost = 20

    if country != local_country:
        cost += 15

    if premium:
        cost += 25

    return cost

assert shipping_cost(0.5, "IN") == 5
assert shipping_cost(3, "KR") == 25
assert shipping_cost(6, "KR") == 35
assert shipping_cost(2, "IN", premium=True) == 35
````

````python
# Pick earliest valid date
def earliest_valid(dates):
    valid_dates = []
    
    for y, m, d in dates:
        if 1 <= m <= 12 and 1 <= d <= 31:
            valid_dates.append((y, m, d))
    
    if not valid_dates:
        return None
    
    return min(valid_dates)
    
assert earliest_valid([(2024,3,10),(2023,15,1),(2024,1,5)]) == (2024,1,5)
assert earliest_valid([(2024,15,1),(2023,13,2),(2024,0,5)]) is None
````

````python
# Safe cast
def to_int(s, default=None):
    s = s.strip()
    
    if s.isdigit():
        return int(s)
    
    return default

assert to_int(" 12 ") == 12
assert to_int("12.0") is None
````

````python
# Guard clauses
def discount(price, code):
    if price <= 0:
        return 0
    
    if code == "NONE":
        return price
    
    if code == "HALF":
        return price * 0.5
    
    return price * 0.9

assert discount(100, "NONE") == 100
assert discount(100, "HALF") == 50
assert discount(100, "VIP") == 90
````

````python
# Ternary map
def label_numbers(nums):
    return [
        "neg" if n < 0 else "zero" if n == 0 else "pos"
        for n in nums
    ]
    
assert label_numbers([-1,0,2]) == ["neg","zero","pos"]
````
