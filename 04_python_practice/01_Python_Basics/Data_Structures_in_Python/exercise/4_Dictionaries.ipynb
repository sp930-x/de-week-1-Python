````python
# Make profile with defaults
def make_profile(name, email=None, country="IN"):
    return {
        "name": name,
        "email": email,
        "country": country,
    }

p = make_profile("A","a@example.com")
assert p["name"]=="A" and p["country"]=="IN"
````

````python
# Safe get path
def get_in(d, path, default=None):
    cur = d
    for key in path:
        if not isinstance(cur, dict) or key not in cur:
            return default
        cur = cur[key]
    return cur
    
cfg = {"db":{"host":"localhost","port":5432}}
assert get_in(cfg, ["db","port"]) == 5432
assert get_in(cfg, ["db","user"], "postgres") == "postgres"
````

````python
# Set with default container
def multimap_add(mm, key, value):
    mm.setdefault(key, []).append(value)
    
mm = {}
multimap_add(mm,"a",1); multimap_add(mm,"a",2)
assert mm == {"a":[1,2]}
````

````python
# Invert mapping (values unique)
def invert_unique(d):
    result = {}
    for k, v in d.items():
        if v not in result:
            result[v] = k
    return result

assert invert_unique({"a":1,"b":2}) == {1:"a",2:"b"}
````

````python
# Invert to sets (values may repeat)
def invert_to_sets(d):
    result = {}
    for k, v in d.items():
        result.setdefault(v, set()).add(k)
    return result
    
assert invert_to_sets({"a":1,"b":1,"c":2}) == {1:{"a","b"}, 2:{"c"}}
````

````python
# Merge with priority
def merge_priority(base, override):
    out = base.copy()
    out.update(override)
    return out
    
assert merge_priority({"a":1,"b":2},{"b":9,"c":3}) == {"a":1,"b":9,"c":3}
````

````python
# Group by key function
def group_by(xs, key_fn):
    result = {}

    for x in xs:
        k = key_fn(x)
        result.setdefault(k, []).append(x)

    return result
    
g = group_by(["apple","pear","plum","kiwi"], len)
assert g == {5:["apple"],4:["pear","plum","kiwi"]}
````

````python
# Top-N by value
def top_n(d, n):
    result = []

    for k, v in d.items():
        result.append((k,v))

    result.sort(key=lambda kv: kv[1], reverse=True)

    return result[:n]
    
assert top_n({"a":3,"b":1,"c":5},2) == [("c",5),("a",3)]
````

````python
# Join two “tables” by key
def left_join(left_rows, right_rows, key):
    # create index
    right_index = {}
    for row in right_rows:
        k = row[key]
        if k not in right_index:
            right_index[k] = row

    out = []
    for left_row in left_rows:
        merged = left_row.copy()
        match = right_index.get(left_row[key])
        if match is not None:
            merged.update(match)
        out.append(merged)
    return out

L = [{"id":1,"name":"A"},{"id":2,"name":"B"}]
R = [{"id":1,"score":90}]
out = left_join(L,R,"id")
assert out[0]["score"]==90 and out[1].get("score") is None
````

````python
# Flatten nested dict with dot paths
def flatten_dot(d, parent_key=""):
    result = {}

    for k, v in d.items():
        new_key = f"{parent_key}.{k}" if parent_key else k

        if isinstance(v, dict):
            result.update(flatten_dot(v, new_key))
        else:
            result[new_key] = v

    return result
    
assert flatten_dot({"a":{"b":1},"c":2}) == {"a.b":1,"c":2}
````
