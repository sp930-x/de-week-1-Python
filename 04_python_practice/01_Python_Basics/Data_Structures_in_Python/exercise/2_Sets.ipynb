````python
# Unique words (case-insensitive)
def unique_words(s):
    result = set()
    
    for word in s.split():
        w = word.lower()
        if w not in result:
            result.add(w)
            
    return result
    
assert unique_words("Red red BLUE blue") == {"red","blue"}
````

````python
# Is pangram
def is_pangram(s):
    letters = {ch.lower() for ch in s if ch.isalpha()}
    return len(letters) == 26
    
assert is_pangram("The quick brown fox jumps over a lazy dog")
````

````python
# Disjoint tags
def disjoint(a_tags, b_tags):
    return len(a_tags & b_tags) == 0
    
assert disjoint({"sql","python"}, {"go"}) is True
assert disjoint({"sql","python"}, {"python","go"}) is False
````

````python
# Set algebra helper
def set_ops(a, b):
    return {
        "union": a | b,
        "intersection": a & b,
        "diff_a_b": a - b,
        "diff_b_a": b - a,
        "symdiff": a ^ b,
    }
    
d = set_ops({1,2,3},{3,4})
assert d["union"] == {1,2,3,4} and d["symdiff"] == {1,2,4}
````

````python
# Seen-first dedupe
def dedupe_keep_first(xs):
    out, seen = [], set()
    for x in xs:
        if x not in seen:
            seen.add(x)
            out.append(x)
    return out
    
assert dedupe_keep_first([1,2,1,3,2,4]) == [1,2,3,4]
assert dedupe_keep_first([1,2,1,3,2,4]) == [1,2,3,4]
````

````python
# Build inverted index
def invert_index(items):
    result = {}
    for item_id, tags in items:
        for tag in tags:
            if tag not in result:
                result[tag] = set()
            result[tag].add(item_id)
    return result
    
ii = invert_index([(1,{"a","b"}),(2,{"b"}),(3,{"a"})])
assert ii == {"a":{1,3}, "b":{1,2}}
````

````python
# Undirected edges as frozensets
def canonical_edge(a, b):
    return frozenset((a, b))
    
assert canonical_edge("A","B") == frozenset({"A","B"})
````

````python
# Missing smallest positive
def first_missing_positive(xs):
    positive_int={x for x in xs if x > 0}
    answer=1
    while answer in positive_int:
        answer+=1
    return answer
    
assert first_missing_positive([3,4,-1,1]) == 2
assert first_missing_positive([1,2,0]) == 3
````

````python
# Triangle check in an undirected graph
from collections import defaultdict

def has_triangle(edges):
    adj = defaultdict(set)

    for u, v in edges:
        adj[u].add(v)
        adj[v].add(u)

    for u in adj:
        for v in adj[u]:
            common = adj[u] & adj[v]
            if common - {u, v}:
                return True
    return False
    
assert has_triangle([("a","b"),("b","c"),("c","a")]) is True
assert has_triangle([("a","b"),("b","c")]) is False
````
